```python
import PyPDF2
```

```python
with open("/Users/rajeshprabhakarkaila/Downloads/RIL_IAR_2024.pdf",'rb') as file:
    reader=PyPDF2.PdfReader(file)
    text=""
    for page_num in range(len(reader.pages)):
        page=reader.pages[page_num]
        text+=page.extract_text()
```

```python
leadership_terms = ['CEO', 'Chief Executive Officer', 'Board of Directors', 'Chairman',
                    'Executive', 'Senior Management', 'Leadership', 'Management Team', 
                    'Chief Officer', 'President','Vice President', 'COO', 'CFO', 'CIO', 
                    'Leadership team', 'Director', 'Manager']
```

```python
import re

# Function to find sentences with leadership and management-related terms
def find_leadership_sections(text, terms):
    # Split text into sentences using regular expression for sentence boundary
    sentences = re.split(r'(?<=[.!?]) +', text)
    # Find sentences that contain any of the terms (case-insensitive)
    relevant_sentences = [sentence for sentence in sentences if any(term.lower() in sentence.lower() for term in terms)]
    return relevant_sentences

# Get sentences that mention leadership/management terms
leadership_sentences = find_leadership_sections(text, leadership_terms)

# Print out the sentences found
for sentence in leadership_sentences:
    print(sentence)
```
